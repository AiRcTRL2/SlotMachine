```python
import emoji
from random import choice, shuffle

faces = [
    emoji.emojize(':red_apple:'),
    emoji.emojize(':pear:'),
    emoji.emojize(':tangerine:')
]
```

```python
class Purse():  # Create Purse class, initiates a balance and has methods for altering the balance depending on win or loss
                # of the Slot class
    def __init__(self, balance=10): # Initiate balance with 10 if none passed as parameter to Purse()
        self.balance = balance
        
    def debit(self, winLoss):
        self.balance = self.balance - winLoss
    
    def credit(self, winLoss):
        self.balance = self.balance + winLoss
    
    def get_balance(self): # Return current balance
        return self.balance
```

```python
class Column(): 
    # Class which initiates a face which is represented by a fruit.
    # Has a change_face method that changes the fruit on face using random.choice module when pull_handle()) is invoked from
    # Slot()
    
    def __init__(self):
        self.change_face()
        
    def change_face(self):
        self.face = choice(faces)
        
    def get_face(self): # Return current value of face.
        return self.face
```

I chose to use an array for each column object instead of setting them up as separate entities. Though I do assign their index to 3 separate variables later, this was easier during the testing phase. I recognise it probably makes more sense to set each Column() object up separately as c1, c2 and c3 and access their values using each object. It would have saved the need for assigning the index to variables x,y,z in the score_slot() method, or index checking in the show_slot() method.. It would be a quick change, but I opted to leave it this way.

```python
class Slot():
    p = Purse() # Initiate purse object so we can add/subtract based on Slot machine results. If we call Purse() in the calculations
                # later, Purse() would reset the value with each call of the function.
    def __init__(self): # Initialise an instance variable with an array that calls Column() three times. 
        self.faces = [Column(), Column(), Column()]
        
    def pull_handle(self):  # When user plays the game, call change_face() method in Column() to get new faces for each self.face
                            # in the Column() object.
        for c in self.faces:
            c.change_face()
    
    def show_slot(self): # Return the self.face value from each column object in self.faces
        return self.faces[0].face,self.faces[1].face,self.faces[2].face
    
    def take_bet(self, bet=2): # Set a bet. If none provided, default bet is 2.
        self.bet = bet
        return self.bet # So we can access the bet value in score_slot()
        
    def score_slot(self): # Method to calculate winnings or losses
        self.bet_result = self.bet # Assign self.bet_result with the value of get_bet().self.bet
        x = self.faces[0] # Assign each column object to a variable.
        y = self.faces[1]
        z = self.faces[2]

        if (x.face == y.face and x.face == z.face): # Checks the .face value in each object. Calculates winnings based on 
                                                    # comparing the .face of each object.
            self.prize = self.bet_result * 2
            p.credit(self.prize) # Call the credit method in Purse()
            print("You won double your money! Winnings: ", self.prize)
            
        elif (x.face == y.face and y.face != z.face) or (y.face == z.face and x.face != y.face) or (x.face == z.face and y.face != z.face):
            self.prize = self.bet_result + (self.bet_result/2)
            p.credit(self.prize)
            print("You won your money back and half of your bet value. Winnings: ", self.prize)
            
        else:
            self.prize = self.bet_result
            p.debit(self.prize) # Call the debit method in Purse()
            print("Sorry, you didn't win this time. Try again! Good luck.")
            
       # return self.prize ## Unnecessary unless we would like to check the value of a particular prize winning instance.
```

```python
p = Purse()
s = Slot()
```

This function was designed with the idea that we don't only want to have 1 purse object and 1 slot object. So, I have allowed for the run_slot_machine() function to take two parameters, which should be previously declared Purse() and Slot() objects. i.e. We could create two new objects p1 and s1, and call the run_slot_machine(p1,s1). This allows for many different players or different slot machines with their own save states. 

```python
def run_slot_machine(p,s):
    """ This function is designed to run our slot machine using objects and classes by taking two parameters, which are 
        objects that must be defined before calling the function.
        
        It assumes that user inputs are whole numbers, even though the machine will return a float as prize money. The machine
        will exit if the user tries to credit their balance with a float number (only accepts integers), or if the user tries
        to bet something other than a whole number, they will receive an error. They may receive this error twice, which will
        trigger the program to exit. They may continue by calling this function again (balance will not be depleted since our
        objects are declared outside the function.)
    """
    print("============== Slot Machine ==============")
    print()
    print("Minimum bet is 2. Type 'N' to exit.")
    print()
    bet_or_quit = input("Enter the amount you wish to bet: ")               # Accept string input, since we need to check
                                                                            # for 'N'.
    print()
    
    while bet_or_quit != "N": # Check for 'N'
        try: # Try to convert input to an integer
            take_bet = int(bet_or_quit) 
        except: # If not possible, explain error to user. Allow them to try again. If fails again, exit program.
            print("Sorry. You attempted to enter a non-whole number. Please ensure you are entering an integer.")
            bet_or_quit = input("Enter the amount you wish to bet (Minimum bet is 2): ")
            try:
                take_bet = int(bet_or_quit)
            except:
                print("Another non-integer was entered. The game will now exit. Please run again to continue playing.")
                return

        if take_bet >=2 and take_bet <= p.get_balance(): # Check if the bet is within the min bet and available balance
            s.take_bet(take_bet)
            s.pull_handle()
            print("============== Handle Pulled ==============")
            print()
            print(s.show_slot())
            print()
            s.score_slot()
            print()
            print("Your balance is now: ", p.get_balance())
            print()
            x = p.get_balance()
            if x == 0.0: # After bet has completed, check balance. If balance reaches 0, ask user if they would like to top up.
                print("Oh no! Your balance has reached 0. Would you like to enter more? If so, insert below. If not, type 'N' to exit.")
                add_or_quit = input("Enter the amount you would like to add to the game (Your card will be debited for this amount): ")
                print("===========================================")
                if add_or_quit == "N": # Allow for string 'N' again. Convert to INT if valid. Exit if N. Will also exit on fail.
                    print("You have chosen to stop playing. We hope to see you soon! Thanks for playing.")
                else:
                    try:
                        credit_balance = int(add_or_quit)
                    except:
                        print("You have tried to add a non-whole number, or have entered a non-integer. Game will now exit. Please restart if you would like to continue.")
                print("You have topped up by: ",credit_balance)
                p.credit(credit_balance) # Add balance to class object.
        elif take_bet <2: # Check if bet is below min. If true, tell user and continue loop to take another input.
            print("Sorry. Minimum bet is 2. Please try again.")
            print("===========================================")
        elif take_bet >p.get_balance(): # Check if attempted bet is greater than available balance and prompt for top-up.
            print("Sorry. Insufficient funds. Would you like to add more?")
            print()
            addMore = input("Enter the amount you'd like to add to your purse (or type 'N' to quit): ") # Allow string again
            if addMore != "N": # Check string
                try:
                    credit_purse = int(addMore) # Convert string
                except: 
                    print("Could not add value to purse. Input may not have been an integer. Exiting...")
                    return
            p.credit(credit_purse) # Add to class object
            print("Your balance is now", p.get_balance())
            print()
            print("===========================================")
        bet_or_quit = input("Enter the amount you wish to bet: ")                    # Get new value for bet_or_quit before
                                                                                     # retarting our loop.
    else:
        print("You have chosen to exit the game. Thank you for playing.")
        return
```

```python
run_slot_machine(p,s)
```
